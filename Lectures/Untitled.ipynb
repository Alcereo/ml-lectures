#### Линейные методы классификации

Решается задача классификации с двумя классами: $Y=\{y^1,y^2\}=\{-1,+1\}$.  
Данные представляюстя в виде $X=\{x_1,x_2,x_3,...,x_l\}$ - множество векторов $x$ с признаками $f_j$, $x=(x^1,x^2,x^3,...,.x^4)=(f_1(x),f_2(x),...,f_n(x))$. Каждому вектору в соответствии его класс $\{y_1,y_2,y_3,...,y_l\} \in Y=\{-1,1\}$ 

---

Вводится $F(x,w)$ - **дискриминантная функция**, которая:

- $F(x,w)>0 =>$ $x$ - относится к классу - $y^2=+1$
- $F(x,w)<0 =>$ $x$ - относится к классу - $y^1=-1$

Нужно найти значение вектора $w$ по обучающей выборке.

---

Вводится функция $M_i(w)=y_i F(x_i,w)$ - **отступ (margin)** объекта $x_i$.  
Очевидно, что $M_i(w)$ - тем больше, чем правильнее и надежнее классификация объекта $x_i$ вектором $w$ и если $M_i(w)<0$ значит объект был классифицирован не верно.

---

Определяется **функция потерь** $L(M_i(w))$ - монотонно возрастающая функция отступа - мажорирующая пороговую функцию потерь.

Теперь задачу поиска можно обозначить как минимизацию функции потерь - $Q$.
$$\min_{w}Q(w, X)=\min_{w}\sum_{i=1}^{l}L(M_i(w))$$

---

#### Линейный классификатор

Случай, когда $F$ - дискриминантная функция описывается как скалярное произведение векторов $x$ и $w$:

$$ F=\langle w,x \rangle - w_0$$

Уравнение $\langle w,x \rangle=0$ задает гиперплоскость, разделяющую классы. Если вектор находится по одну сторону с направляющим вектором $w$, то относится к классу $y_2=+1$, иначе $-1$

---

#### Пример с 2х мерном случае

$X_1, X_2$ - два множества 2-х мерных векторов.  
$W^1$ - вектор коэффициентов  
$W^0$ - коэффициент $w_0$  

```python
X_1 = np.array([[np.random.randint(10,60), 
                 np.random.randint(40,80)] 
                for n in range(10)])

X_2 = np.array([[np.random.randint(60,80), 
                 np.random.randint(10,40)] 
                for n in range(10)])

W_1 = np.array([
    [0,0],
    [-1,1]
])

W_0 = -20
```

```python
from __future__ import print_function, division
import numpy as np
%matplotlib inline
import matplotlib as mpl
# mpl.use("WebAgg")  # Must come after importing mpl, but before importing plt
import matplotlib.pyplot as plt
```

```python
# mpl.style.use('seaborn')
mpl.style.use('classic')

plt.figure(figsize=(15,10));
ax = plt.axes()
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
# ax.plot();
plt.ylim([0,100])
plt.xlim([-30,100])
plt.plot(X_1[:,0], X_1[:,1],".",color="red"); # X_1
plt.plot(X_2[:,0], X_2[:,1],".",color="blue"); # X_2

# Разделяющая прямая
plt.plot([20, 120],[0,100]);
plt.annotate('прямая разделяющая \n      классы',xy=(90,70), xytext=(55,85),
             arrowprops=dict(facecolor="grey")
            )

# Вектор параметров
plt.arrow(40,20,
          -12,15, 
          head_width=1.5, 
          head_length=2, 
          color="k"
         );

plt.annotate('$W^1$',
              xy=(40,20), 
             xytext=(37,27),
            )

# Поправочный параметр
plt.plot([0,0],[0,10], color='black')
plt.plot([20,20],[0,10], color='k')

plt.annotate("",
             xy=(0,8),
             xytext=(20,8),
             arrowprops=dict(
                 arrowstyle="<->",
                 facecolor="grey"
             )
            );

plt.annotate("$W^0$",
             xy=(10,8),
             xytext=(10,9),
            );

# plt.show();
```

$W=(-1,1)$  
$W^0=-20$  
$F=\langle (x^1, x^2), (-1, 1) \rangle + W^0 = x^2-x^1+20$
